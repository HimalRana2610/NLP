```python
import matplotlib.pyplot as plt
```

```python
with open("tokenized_sentences_from_words.txt", "r", encoding="utf-8") as f:
    tokens = f.read().split()
freq_dist = {}
for word in tokens:
    if word not in freq_dist:
        freq_dist[word] = 1
    else:
        freq_dist[word] += 1
sorted_freq = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
top_100 = sorted_freq[:100]
words, counts = zip(*top_100)
plt.figure(figsize=(20,6))
plt.bar(words, counts)
plt.xticks(rotation=90)
plt.title("Top 100 Most Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
def remove_stopwords(freq_dict, threshold):
    stopwords = {w for w, c in freq_dict.items() if c >= threshold}
    new_freq = {w: c for w, c in freq_dict.items() if w not in stopwords}
    return new_freq, stopwords
```

```python
thresholds = [50, 100, 200]
for t in thresholds:
    filtered_freq, stopwords = remove_stopwords(freq_dist, t)
    sorted_filtered = sorted(filtered_freq.items(), key=lambda x: x[1], reverse=True)[:100]
    words, counts = zip(*sorted_filtered)
    plt.figure(figsize=(20,6))
    plt.bar(words, counts)
    plt.xticks(rotation=90)
    plt.title(f"Top 100 Words After Removing Stop Words (Threshold={t})")
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.show()
```
